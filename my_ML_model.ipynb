# Trying my first machine learning model with a recipes csv to predict the country base on their food

### import necessary libraries for the model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
print('Setup complete')
```

### Read the main csv file for the model

```python
recipes_data = pd.read_csv(r"C:\Users\anim6\Downloads\recipes.csv")
recipes_data.columns
```

```python
recipes_data.describe()
```

```python
recipes_data.head()
```

### Changing all Yes to 1 and No to 0 

```python
recipes_data_new = recipes_data.replace({'No': 0, 'Yes': 1})
recipes_data_new.head()
```

```python
recipes_data_new.to_csv('recipes_1.csv', index=False)
```

```python
y = recipes_data_new["country"]

X = recipes_data_new.drop(columns=["country"])
```

### splitting data into train, test and validation sets

```python
Train_X, Val_X, Train_y, Val_y = train_test_split(X, y, random_state=1, test_size=0.2, stratify=y)
```

## The models

### DecisionTreeClassifier

```python
recipe_model_DTC = DecisionTreeClassifier(random_state=40, 
                                         criterion= 'gini', 
                                         max_depth= 10, 
                                         max_features= None, 
                                         min_samples_leaf= 11, 
                                         min_samples_split= 10
                                         )
recipe_model_DTC.fit(Train_X, Train_y)
```

```python
val_predictions = recipe_model_DTC.predict(Val_X)
val_predictions
```

```python
val_cm = confusion_matrix(Val_y, val_predictions)
val_cm
```

```python
accuracy = accuracy_score(Val_y, val_predictions)

print(f"Accuracy: {accuracy:.4f}")
```

```python
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy'],
    'max_features': [None, 'sqrt', 'log2']
}

# Grid search
grid_search = GridSearchCV(estimator=recipe_model_DTC, param_grid=param_grid, cv=3, scoring='accuracy')
grid_search.fit(Train_X, Train_y)

# Best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)
```

### RandomForestClassifier

```python
recipe_model_RFC = RandomForestClassifier(
    n_estimators=100, 
    random_state= 40, 
    min_samples_split=10, 
    min_samples_leaf=2, 
    max_features='sqrt', 
)
recipe_model_RFC.fit(Train_X,Train_y)
```

```python
val_pred = recipe_model_RFC.predict(Val_X)
val_pred
```

```python
Val_CM = confusion_matrix(val_pred, Val_y)
Val_CM
```

```python
ACCURACY = accuracy_score(Val_y, val_pred)
print(f"Accuracy: {ACCURACY:.4f}")
```

```python
from sklearn.model_selection import RandomizedSearchCV

param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 30, 50, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

rf = RandomForestClassifier(random_state=42)
rand_search = RandomizedSearchCV(rf, param_distributions=param_dist,
                                 n_iter=10, cv=3, scoring='accuracy', n_jobs=-1)

rand_search.fit(Train_X, Train_y)

print("Best Accuracy:", rand_search.best_score_)
print("Best Params:", rand_search.best_params_)
```
